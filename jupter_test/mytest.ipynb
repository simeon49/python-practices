# 将由多个单词拼装的文本进行拆分

```python
s = 'catsanddog'
wordDict = {
    'cat': 1,
    'cats': 1,
    'and': 1,
    'sand': 1,
    'dog': 1,
}

def can_text_break(text, wd):
    length = len(text)
    dp = [False] * (length + 1)
    dp[0] = True
    for index in range(length):
        for cur in range(index, -1, -1):
            if dp[cur] and text[cur:index + 1] in wd:
                dp[index + 1] = True
                break
    return dp[length]

def dfs(text, wd, res, tmp):
    if not text:
        res.append(tmp[:])
        return
    for word in wd:
        length = len(word)
        if word == text[:length]:
            tmp.append(word)
            dfs(text[length:], wd, res, tmp)
            tmp.pop()

def word_break(text, wd):
    res = []
    tmp = []
    if not can_text_break(text, wd):
        return []
    dfs(text, wd, res, tmp)
    return res

word_break(s, wordDict)
```
